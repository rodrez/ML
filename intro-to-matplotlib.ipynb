## Matplotlib example workflow

```python
# 0. import matplotlib and get ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Perpare Data
x = [1,2,3,4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #(width, height)
# dimesiones of this plot (x, y)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (show the whole figure)
fig.savefig("sample-plt.png")
```

## Making figures with Numpy arrays
We want:
* Line plot
* Scatter plot
* Bar plot 
* Histogram
* Subplot

```python
import numpy as mp
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same dat to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make from dictionary
nut_butter_price = {"Almond butter": 10,
                    "Peanut butter": 8,
                   "Cahsew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values());
ax.set(title="Nut Butter Store",
       ylabel="Prices ($)");

# semi colon helps to remove the output text
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()), list(nut_butter_price.values()));
# When using horizontal bars we must convert to list
```

```python
# Make some data for histogram and plot it 
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

# Two options 

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                             figsize=(10,5))
# Plot to each differnet axis
ax1.plot(x , x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1, 1].hist(np.random.randn(1000));
```

```python
df["Price"] = df["Price"].str.replace('[\$\.\,]', "")
df
```

```python
# Remove last two zeroes
df["Price"] = df["Price"].str[:-2]
```

```python
df
```

```python
df["Sale Date"] = pd.date_range("1/1/2020", periods=len(df))
df
```

```python
df["Total Sales"] = df["Price"].astype(int).cumsum()
```

```python
df
```

```python
# Lets plot the total sales
df.plot(x="Sale Date", y="Total Sales");
```

```python
#Reassign price column to int
df["Price"] = df["Price"].astype(int)

# Plot scatter plot with price column
df.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
