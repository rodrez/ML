## Matplotlib example workflow

```python
# 0. import matplotlib and get ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Perpare Data
x = [1,2,3,4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #(width, height)
# dimesiones of this plot (x, y)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (show the whole figure)
fig.savefig("sample-plt.png")
```

## Making figures with Numpy arrays
We want:
* Line plot
* Scatter plot
* Bar plot 
* Histogram
* Subplot

```python
import numpy as mp
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same dat to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make from dictionary
nut_butter_price = {"Almond butter": 10,
                    "Peanut butter": 8,
                   "Cahsew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values());
ax.set(title="Nut Butter Store",
       ylabel="Prices ($)");

# semi colon helps to remove the output text
```
