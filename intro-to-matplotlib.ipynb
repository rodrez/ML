## Matplotlib example workflow

```python
# 0. import matplotlib and get ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Perpare Data
x = [1,2,3,4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #(width, height)
# dimesiones of this plot (x, y)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (show the whole figure)
fig.savefig("sample-plt.png")
```

# Two options 

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                             figsize=(10,5))
# Plot to each differnet axis
ax1.plot(x , x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use pyplot or matplotlop OO method?

* When plotting something quick, okay to use pyplot
* When plotting something advance, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot mehtod
over_50.plot(kind="scatter",
            x='age',
            y="chol",
            c="target");
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);

# ax.set_xlim([45, 100])
```

```python
over_50["target"].values
```

```python
# OO Method from sratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap='winter')

# Customize
ax.set(title="Heart Diseasde and Cholesterol Levels",
      xlabel='Age',
      ylabel='Cholesterol')

# add legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Add Legend ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#Add horizontal Line ax0
ax0.axhline(over_50["chol"].mean(),
          linestyle='--');

# Adding data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

#Customizing ax1
ax1.set(title="Heart Disease and Max Heart rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line ax1
ax1.axhline(over_50["thalach"].mean(),
           linestyle='--')


# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set style
plt.style.use("seaborn-whitegrid")

# OO Method from sratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap='winter')

# Customize
ax.set(title="Heart Diseasde and Cholesterol Levels",
      xlabel='Age',
      ylabel='Cholesterol')

# add legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# Customizing y and x axis limitations
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")
# Change X axis limits
ax0.set_xlim([50, 80])

# Add Legend ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#Add horizontal Line ax0
ax0.axhline(over_50["chol"].mean(),
          linestyle='--');

# Adding data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

#Customizing ax1
ax1.set(title="Heart Disease and Max Heart rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

#Change y limitation
ax1.set_ylim(60, 200)

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line ax1
ax1.axhline(over_50["thalach"].mean(),
           linestyle='--')


# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
